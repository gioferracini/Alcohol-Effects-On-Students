# **We're going to start working the data to divide the students by clusters**

```python
#Separating the dataset in 2

X = math_2.iloc[:, :32]
y = math_2.iloc[:, 32:33]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

k_best = SelectKBest(chi2, k = 10)
k_best.fit(X, y)

col = k_best.get_support(1)
math_3 = X[X.columns[col]]
math_3 = math_3.join(y)
math_copy = math_3.copy()
math_copy
```

After we have discovered the best features of dataset, we're going to work it separating the categorical columns, to unique each one to process how will affect between then in the correlation matriz.

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(dtype = int)
address_ohe = ohe.fit_transform(math_3[['address']]).toarray()
math_3 = math_3.drop(columns = 'address', axis = 1)
math_3 = pd.concat([math_3, pd.DataFrame(address_ohe, columns = ohe.get_feature_names_out(['address']))], axis = 1)

Mjob_ohe = ohe.fit_transform(math_3[['Mjob']]).toarray()
math_3 = math_3.drop(columns = 'Mjob', axis = 1)
math_3 = pd.concat([math_3, pd.DataFrame(Mjob_ohe, columns = ohe.get_feature_names_out(['Mjob']))], axis = 1)

failures_ohe = ohe.fit_transform(math_3[['failures']]).toarray()
math_3 = math_3.drop(columns = 'failures', axis = 1)
math_3 = pd.concat([math_3, pd.DataFrame(failures_ohe, columns = ohe.get_feature_names_out(['failures']))], axis = 1)

schoolsup_ohe = ohe.fit_transform(math_3[['schoolsup']]).toarray()
math_3 = math_3.drop(columns = 'schoolsup', axis = 1)
math_3 = pd.concat([math_3, pd.DataFrame(schoolsup_ohe, columns = ohe.get_feature_names_out(['schoolsup']))], axis = 1)

romantic_ohe = ohe.fit_transform(math_3[['romantic']]).toarray()
math_3 = math_3.drop(columns = 'romantic', axis = 1)
math_3 = pd.concat([math_3, pd.DataFrame(romantic_ohe, columns = ohe.get_feature_names_out(['romantic']))], axis = 1)

Dalc_ohe = ohe.fit_transform(math_3[['Dalc']]).toarray()
math_3 = math_3.drop(columns = 'Dalc', axis = 1)
math_3 = pd.concat([math_3, pd.DataFrame(Dalc_ohe, columns = ohe.get_feature_names_out(['Dalc']))], axis = 1)

Walc_ohe = ohe.fit_transform(math_3[['Walc']]).toarray()
math_3 = math_3.drop(columns = 'Walc', axis = 1)
math_3 = pd.concat([math_3, pd.DataFrame(Walc_ohe, columns = ohe.get_feature_names_out(['Walc']))], axis = 1)
```

# **Standardizing the data**

```python
from sklearn.preprocessing import StandardScaler

standardizer = StandardScaler()
standardizer.fit(math_3)
math_4 = standardizer.transform(math_3)
math_4 = pd.DataFrame(data = math_4, columns = math_3.keys())
```

```python
corr = math_4.corr()
graph = sns.heatmap(corr, annot = True, fmt = '.2f')
graph.figure.set_size_inches(24, 16)
graph.set_title('Correlation Matriz', fontsize = 30)
```

# **Identifying the distribution of clusters in a graph**

```python
SEED = 1234
np.random.seed(1234)

from sklearn.decomposition import PCA 

pca = PCA(n_components = 2, random_state = SEED)
math_pca = pca.fit_transform(math_4)
math_pca_info = pd.DataFrame(columns = ['X', 'y'], data = math_pca)
math_pca_info
```

```python
from sklearn.cluster import KMeans

kmeans_pca = KMeans(n_clusters = 5, verbose = True, random_state = SEED)
kmeans_pca.fit(math_pca_info)
math_pca_info['cluster_pca'] = kmeans_pca.predict(math_pca_info)
```

```python
math_pca_info
```

```python
import plotly.express as px

fig = px.scatter(
    math_pca_info, x = 'X', y = 'y', color = 'cluster_pca', hover_data=['X', 'y', 'cluster_pca'] 
)
fig.show()
```

newplot (1).png

```python
math_description = math_copy.join(math_pca_info['cluster_pca'])
math_description
```

# **As the dataset is small, we're going to use the median value to identify how were composed the clusters**

```python
description = math_description.groupby('cluster_pca')['address', 'Mjob', 'failures', 'schoolsup', 'romantic', 'Dalc', 'Walc', 'absences', 'G1', 'G2', 'G3']
n_students = description.size()
description = description.median()
description['n_students'] = n_students
print(description)
```

**Cluster 0:** The cluster with lower grades and presents the highest number of absences and alcohol consumption. Probably the worst math' students of dataset.

**Cluster 1:** The cluster with lower alcohol consumption and highest math's grades. Probably the best math' students of dataset.

**Cluster 2:** The cluster composed from most students. These have a low alcohol consumption, however, the grades are mean and the number of absences are a little troublesome.

**Cluster 3:** The cluster with less number of students. These need support classes because their grades are low even with few alcohol consumption. Maybe there're other reasons besides of alcohol to explain the badly performance.

**Cluster 4:** A curious case. Because equal the cluster 0, this presents a high alcohol consumption, but the grades are good too. Maybe the consumption are big because they are experient students and know the subject.
